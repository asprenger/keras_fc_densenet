Plot CamVid images and labels from TFRecord file.

```python
import os
import sys
sys.path.append(os.path.abspath(os.path.join('..')))

import matplotlib.pyplot as plt
import numpy as np
from camvid_dataset import dataset, load_camvid_tfrecords
from camvid_utils import load_label_colors, color_label, unnormalize_image
from common import show_image_row
```

```python
label_colors_path = '/Users/asprenger/work/camvid/label_colors.txt'
label_colors, _, _ = load_label_colors(label_colors_path)
```

```python
camvid_path = '/tmp/camvid-preprocessed/camvid-384x480-train.tfrecords'
images, labels = load_camvid_tfrecords(camvid_path, 5)
print('Images: %s, %s' % (images.shape, images.dtype))
print('Labels: %s, %s' % (labels.shape, labels.dtype))
```

```python
for i in range(len(images)):
    image = unnormalize_image(images[i])    
    label = labels[i].reshape((image.shape[0],image.shape[1]))
    label = color_label(label, label_colors)
    show_image_row([image, label])
    plt.show()
```
